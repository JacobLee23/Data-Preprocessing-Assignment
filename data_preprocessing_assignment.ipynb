```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def preprocess_data(path: str, summary: bool = False) -> pd.DataFrame:
    """
    Parses the GenePattern GCT file (.gct) at ``path`` into a ``DataFrame``.
    
    Parameter ``path`` is assumed to be a valid filepath or URL address to a properly formatted GCT
    file. Neither ``path`` nor the contents of the file at ``path`` are validated.
    
    .. note::
    
        The first two lines of the input GCT file are disregarded.
    
    :param path: A path or URL address to a GCT file
    :param summary: If ``True``, prints the number rows and columns of the returned ``DataFrame``
    :return: A ``DataFrame`` containing the contents of the GCT file defined by ``path``
    """
    dataframe = pd.read_table(path, delimiter="\t", header=2, index_col=[0, 1])
    
    if summary:
        print(f"Rows: {dataframe.shape[0]}\nColumns: {dataframe.shape[1]}")
        
    return dataframe
```

```python
# 2a. GCT file (60 x 19)

df_2a = preprocess_data("https://datasets.genepattern.org/data/test_data/BRCA_minimal_60x19.gct", summary=True)
```

```python
df_2a.head()
```

```python
# 2b. GCT file (20783 x 40)

df_2b = preprocess_data("https://datasets.genepattern.org/data/test_data/BRCA_large_20783x40.gct")
```

```python
df_2b.head()
```

```python
xscale = 50000
sample_names = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]
subplot_colors = ["tab:blue", "tab:orange", "tab:green", "tab:red"]
```

```python
# 3. Histograms of gene expression values of select columns of DataFrame (2b)

figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
figure.suptitle(
    "Frequency of Gene Expression Values in Select Samples", fontsize=24
)

for axis, column, color in zip([a for x in axes for a in x], sample_names, subplot_colors):
    axis.hist(df_2b.loc[:, column], bins=range(0, 1000000 + xscale, xscale), log=True, color=color)
    axis.set_xticks(range(0, 1000000 + 2 * xscale, 2 * xscale))
    axis.set_title(f"Sample {column}")
    axis.set_xlabel("Gene Expression Value")
    axis.set_ylabel("Frequency")
```

```python
# 4. Descriptive statistics of the expression values for each column in DataFrame (2b)

data = {
    "Sample Name": df_2b.columns,
    "Mean": df_2b.mean(),
    "Median": df_2b.median(),
    "Standard Deviation": df_2b.std()
}

dataframe = pd.DataFrame(data)
dataframe.reset_index(drop=True, inplace=True)
dataframe
```

```python
# 5a. Remove from DataFrame (2b) rows whose maximum value is less than 500
df_5a = df_2b.loc[~(df_2b.max(axis=1) < 500), :]
df_5a.head()
```

```python
# 5b. Descriptive statistics of expression values for each column of DataFrame (5a)

data = {
    "Sample Name": df_5a.columns,
    "Mean": df_5a.mean(),
    "Median": df_5a.median()
}

dataframe = pd.DataFrame(data)
dataframe.reset_index(drop=True, inplace=True)
dataframe
```

```python
# 5b. Histograms of gene expression values of select columns of DataFrame (5a)

figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
figure.suptitle(
    "Frequency of Gene Expression Values in Select Samples (Excluding Genes with Maximum Gene Expression Value less than 500)",
    fontsize=16
)

for axis, column, color in zip([a for x in axes for a in x], sample_names, subplot_colors):
    axis.hist(df_2b.loc[:, column], bins=range(0, 1000000 + xscale, xscale), log=True, color=color)
    axis.set_xticks(range(0, 1000000 + 2 * xscale, 2 * xscale))
    axis.set_title(f"Sample {column}")
    axis.set_xlabel("Gene Expression Value")
    axis.set_ylabel("Frequency")
```
